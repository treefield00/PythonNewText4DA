## ▶︎4.3.3 グラフの種類と出力方法

### ●折れ線グラフ

```
# plotメソッド 折れ線グラフ
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

```
# plot（折れ線）で、擬似的に曲線グラフを表示
import numpy as np

# x[0~15)
x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()
plt.show()

print(x)
print(y1)
```

### ●棒グラフ

```
# barメソッド

fig, ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]
# 棒グラフを描画
ax.bar(x,y)
plt.show()
```

```
# tick_label引数

fig, ax = plt.subplots()
x = [1,2,3]
y = [10,2,3]
labels = ['spam','ham','egg']
# ラベルを指定
ax.bar(x, y, tick_label=labels)
plt.show()
```

```
# 横向き棒グラフ
# barhメソッド

fig, ax = plt.subplots()
x = [1,2,3]
y = [10,2,3]
labels = ['spam','ham','egg']
# ラベルを指定
ax.barh(x, y, tick_label=labels)
plt.show()
```

```
# 複数の棒グラフ
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ['spam', 'ham', 'egg']
# 棒グラフの幅を0.4にする
width = 0.4
# 幅を指定
ax.bar(x,y1, width=width, tick_label=labels, label='y1')
# 幅分ずらす
x2 = [num + width for num in x]
ax.bar(x2,y2, width=width, label='y2')
ax.legend()
plt.show()
```

```
# 積み上げグラフ
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ['spam', 'ham', 'egg']
# y1とy2を足した値を格納
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]
# y1とy2を足した高さのグラフ描画
ax.bar(x,y_total, tick_label=labels, label='y1')
ax.bar(x,y2, label='y2')
ax.legend()
plt.show()
```

### ●ヒストグラム

```
# histメソッド

np.random.seed(123)
mu = 100 #平均
sigma = 15 #標準偏差
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x)
plt.show()
```

```
print(n)
print(bins)
print(patches)
print(patches[0])
print(patches[1])
```

```
# n:各ビンの度数
# bins:ビンの境界の値
# patches:ビンを描画するための情報

for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i+1], num))

print('len(n):',len(n))
print('len(bins):',len(bins))
```

```
# histメソッドのbins引数
# デフォルトビン数は　１０

fig, ax = plt.subplots()
# ビンの数を指定
ax.hist(x, bins=25)
plt.show()
```

```
# orientation引数
# 横向き　orientation='horizontal'
fig, ax = plt.subplots()

ax.hist(x, orientation='horizontal')
plt.show()
```

```
# 複数の値
import numpy as np

np.random.seed(123)
mu = 100
x0 = np.random.normal(mu, 20, 1000)
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)
fig, ax = plt.subplots()
labels = ['x0','x1','x2']

ax.hist((x0,x1,x2), label=labels)
ax.legend()
plt.show()
```

```
# stacked引数
# stacked=True

fig, ax = plt.subplots()
labels = ['x0','x1','x2']
ax.hist((x0,x1,x2), label=labels, stacked=True)
ax.legend()
plt.show()
```

### ●円グラフ

```
# pieメソッド

labels = ['spam', 'ham', 'egg']
x = [10,3,1]

fig, ax = plt.subplots()
# 円グラフ
ax.pie(x, labels=labels)
plt.show()
```

```
# アスペクト比

fig, ax = plt.subplots()
ax.pie(x, labels=labels)
# アスペクト比を保持して描画
ax.axis('equal')
plt.show()
```

```
# デフォルトでは、3時の位置から反時計回りに描画
# 12時の位置　startangle=90
# 時計回り　counterclock=False

fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False)
ax.axis('equal')
plt.show()
```

```
# 影　shadow=True
# パーセント　autopct='%1.2f%%'

fig, ax = plt.subplots()
ax.pie(x,labels=labels, startangle=90,
       counterclock=False,
       shadow=True,
       autopct='%1.2f%%')
ax.axis('equal')
plt.show()
```

```
# explode引数　切り出し

explode = [0, 0.2, 0.3]

fig, ax = plt.subplots()
ax.pie(x,labels=labels, startangle=90,
       counterclock=False,
       shadow=True,
       autopct='%1.2f%%',
       explode=explode
       )

ax.axis('equal')
plt.show()
```

```
# ヒストグラムと近似曲線

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
x = np.random.randn(1000) # 正規乱数を生成

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = plt.hist(x, bins=25)

# 近似曲線に用いる点を求める（ヒストグラムのビンの中点）
x_fit = (edges[:-1] + edges[1:]) / 2
# 近似線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit ** 2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)
plt.show()
```

### ●線のスタイル

```
# 線の幅を指定
# linewidth引数
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot([1,3],[3,1], linewidth=5.5, label='5.5')
ax.plot([1,3],[1,3], linewidth=10, label='10')
ax.legend()
plt.show()
```

```
# 線の種類を指定
# linestyle引数
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
# 破線
ax.plot([1,3],[3,1], linestyle='--', label='dashed')
# 一点鎖線
ax.plot([1,3],[1,3], linestyle='-.', label='dashdot')
# 点線
ax.plot([1,3],[2,2], linestyle=':', label='dotted')
ax.legend()
plt.show()
```

### ●フォント

```
# フォントのスタイルを指定
# タイトル、凡例、軸ラベル
# size引数
# weight引数
# family引数 serif, sans-serif
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_xlabel('xlabel', family='fantasy', size=20, weight='bold')
ax.set_ylabel('ylabel', family='cursive', size=40, weight='light')
ax.set_title('graph title', family='monospace', size=25, weight='heavy')
plt.show()
```

```
# フォントスタイルを辞書形式で指定
# fontdict引数
import matplotlib.pyplot as plt
fontdict = {
    'family':'fantasy',
    'size':20,
    'weight':'normal'
}
fig, ax = plt.subplots()
ax.set_xlabel('x_label', fontdict=fontdict)
ax.set_ylabel('y_label', fontdict=fontdict)
ax.set_title('graph title', fontdict=fontdict, size=40)
plt.show()
```
